# Water Gun Snake Game

This notebook implements a simple game of Water Gun Snake against the computer. It's a variation of the classic Rock-Paper-Scissors game, where players choose between Water(w), Gun(g), and Snake(s).


## Game Rules

Welcome to Water Gun Snake!

Here's how the game works:

*   **Water** drowns **Gun** (Water wins against Gun)
*   **Gun** shoots **Snake** (Gun wins against Snake)
*   **Snake** drinks **Water** (Snake wins against Water)

If both players choose the same item, it's a draw!

## Game Logic Refinement

The game is implemented using a simple rock-paper-scissors-like logic, where user choices ('w' for water, 'g' for gun, 's' for snake) are mapped to numerical values (-1, 0, 1 respectively). The computer's choice is randomly generated from these same numerical values.

The core of the game's win/loss determination lies in analyzing the difference between the computer's choice and the player's choice.

- If `computer == you`: It's a draw.

- Otherwise, the game evaluates `computer - you`:
    - You win if `computer - you` results in specific values (e.g., `-2` or `1`).
        - For instance, if `computer` is water (-1) and `you` are snake (1), then `computer - you` is `-1 - 1 = -2`, which is a win for the player.
        - If `computer` is gun (0) and `you` are water (-1), then `computer - you` is `0 - (-1) = 1`, which is a win for the player.
    - You lose if `computer - you` results in other specific values (e.g., `-1` or `2`).
        - For instance, if `computer` is water (-1) and `you` are gun (0), then `computer - you` is `-1 - 0 = -1`, which is a loss for the player.
        - If `computer` is snake (1) and `you` are water (-1), then `computer - you` is `1 - (-1) = 2`, which is a loss for the player.

This numerical difference approach provides a concise way to determine the outcome, abstracting the individual win/loss conditions into a more generalized logic. The previous conditional statements can indeed be shortened by checking if `computer - you` matches the winning differences.
